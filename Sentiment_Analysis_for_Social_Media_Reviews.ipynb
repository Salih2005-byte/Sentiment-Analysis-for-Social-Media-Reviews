```python
from sklearn.utils import resample

# Get the counts of each sentiment
sentiment_counts = df['Sentiment'].value_counts()

# Find the minority class count (Neutral in this case)
min_count = sentiment_counts['Neutral']

# Separate majority and minority classes
df_negative = df[df['Sentiment'] == 'Negative']
df_positive = df[df['Sentiment'] == 'Positive']
df_neutral = df[df['Sentiment'] == 'Neutral']

# Undersample the majority classes
df_negative_undersampled = resample(df_negative,
                                    replace=False,    # sample without replacement
                                    n_samples=min_count, # to match minority class
                                    random_state=42)  # reproducible results

df_positive_undersampled = resample(df_positive,
                                    replace=False,    # sample without replacement
                                    n_samples=min_count, # to match minority class
                                    random_state=42)  # reproducible results

# Combine minority class with undersampled majority classes
df_balanced = pd.concat([df_negative_undersampled, df_positive_undersampled, df_neutral])

# Display new class counts
print("Original sentiment counts:")
print(sentiment_counts)
print("\nBalanced sentiment counts:")
print(df_balanced['Sentiment'].value_counts())

# Update the main DataFrame to the balanced one
df = df_balanced.reset_index(drop=True)
```

### Preprocessing

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download NLTK data (if not already downloaded)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Extract the key words for sentiment from the stopwords list
stop_words = stop_words - {'no', 'not', 'nothing', 'nor', 'never', 'neither', 'nobody', 'none', 'nowhere'}

def preprocess_text(text):
    if not isinstance(text, str):
        return ""
    # 1. Lowercasing
    text = text.lower()
    # 2. Remove punctuation
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)
    # 4. Tokenization
    tokens = nltk.word_tokenize(text)
    # 5. Remove stop words and Lemmatize
    processed_tokens = [
        lemmatizer.lemmatize(word) for word in tokens if word not in stop_words and word.isalpha()
    ]
    # 6. Join tokens back into a string
    return " ".join(processed_tokens)
```

```python
df['Tweet Text'] = df['Tweet Text'].apply(preprocess_text)
df[['Tweet ID', 'Entity', 'Sentiment', 'Tweet Text']].head()
```

### Modelling

```python
# Split data into X and y
X

y = df["Sentiment"]
```

```python
from sklearn.model_selection import train_test_split
# Split data into train and test sets
np.random.seed(42)

# The train/test split lines are removed to revert the notebook state.
# If you wish to perform the split, please add these lines back:
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

# Put models in a dictionary
models = {"Logistic Regression": LogisticRegression(class_weight='balanced'), # Correct Class Imbalance
          "KNN": KNeighborsClassifier()}

# Create a function to fit and score models
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Fits and evaluates given machine learning models.
    models : a dict of differetn Scikit-Learn machine learning models
    X_train : training data (no labels)
    X_test : testing data (no labels)
    y_train : training labels
    y_test : test labels
    """
    # Set random seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    # Loop through models
    for name, model in models.items():
        # Fit the model to the data
        model.fit(X_train, y_train)
        # Evaluate the model and append its score to model_scores
        model_scores[name] = model.score(X_test, y_test)
    return model_scores
```

```python
model_scores = fit_and_score(models=models,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)

model_scores
```

```python
from sklearn.model_selection import cross_val_score

for model_name, model_estimator in models.items():
  scores = cross_val_score(
    model_estimator,
    X_train,
    y_train,
    cv=5,
    scoring="accuracy"
  )
  print(f"Cross-validation scores for {model_name}: {scores.mean():.4f} (std: {scores.std():.4f})")
```

```python
def predict_sentiment_proba(text, model, vectorizer, preprocess_func):
    # Preprocess the text using the existing function
    preprocessed_text = preprocess_func(text)

    # Vectorize the preprocessed text using the trained vectorizer
    # transformed_text will be a sparse matrix, similar to X
    transformed_text = vectorizer.transform([preprocessed_text])

    # Get probability predictions from the model
    proba = model.predict_proba(transformed_text)[0]

    # Map sentiment classes to probabilities, handling the Turkish labels
    classes = model.classes_ # Get the order of classes from the trained model
    result = {}
    for i, cls in enumerate(classes):
        if cls == 'Negative':
            result['Negative'] = proba[i]
        elif cls == 'Positive':
            result['Positive'] = proba[i]
        elif cls == 'Neutral':
            result['Neutral'] = proba[i]
    return result
```

```python
# Select the best performing model (e.g., Logistic Regression)
best_model = models['Logistic Regression']

# Example usage with a new text
example_text_positive = "This movie was fantastic, absolutely loved it!"
sentiment_probabilities_positive = predict_sentiment_proba(example_text_positive, best_model, vectorizer, preprocess_text)
print(f"Sentiment probabilities for '{example_text_positive}':")
print(sentiment_probabilities_positive)

print("\n" + "-"*50 + "\n")

example_text_negative = "Terrible customer service, very disappointed."
sentiment_probabilities_negative = predict_sentiment_proba(example_text_negative, best_model, vectorizer, preprocess_text)
print(f"Sentiment probabilities for '{example_text_negative}':")
print(sentiment_probabilities_negative)

print("\n" + "-"*50 + "\n")

example_text_neutral = "The new update was released today. Nothing much to say."
sentiment_probabilities_neutral = predict_sentiment_proba(example_text_neutral, best_model, vectorizer, preprocess_text)
print(f"Sentiment probabilities for '{example_text_neutral}':")
print(sentiment_probabilities_neutral)
```

### Cross Validation

```python
from sklearn.model_selection import cross_val_score, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression

# Create a hyperparameter grid for LogisticRegression
log_reg_grid = {"C": np.logspace(-4, 4, 20),
                "solver": ["liblinear"]}

# Tune LogisticRegression

np.random.seed(42)

# Setup random hyperparameter search for LogisticRegression
rs_log_reg = RandomizedSearchCV(LogisticRegression(class_weight='balanced'),
                                param_distributions=log_reg_grid,
                                cv=5,
                                n_iter=20,
                                verbose=True)

# Fit random hyperparameter search model for LogisticRegression
rs_log_reg.fit(X_train, y_train)
```
